タグの定義及び検索ワードの抽出

```
import random
import requests
import lxml.html
from IPython.display import Image
import urllib.parse
% matplotlib inline
import matplotlib.pyplot as plt
import urllib
from skimage import io



class weightscale:
  pass

class behavior:
  pass

class usertime:
  pass

class weather:
  pass

#それぞれのタグの定義

#行動要素
weightscale.activity = ['体重'] 
weightscale.positive = ['健康管理','長生き','美容','イケオジ']
weightscale.negative = ['体調不良','ダイエット','太った','だらしない']

#行動の継続
behavior.activity = ['継続','3日連続']
behavior.positive = ['挑戦','意欲']
behavior.negative = ['飽きた']

#時間の要素
usertime.activity = ['朝','昼','夜']
usertime.season = ['春','夏','秋','冬']
usertime.positive = ['早起き','朝活','いい目覚め','ゆとり生活']
usertime.negative = ['眠い','寝れない','睡眠不足','遅刻']
usertime.others = ['いつもより早い', 'いつもの時間', 'いつもより遅い']

#天気に関わる要素
weather.info = ['晴れ','くもり','雨','PM2.5','通り雨']
weather.feel = ['肌寒い','乾燥している','湿度高い','あったかい']


#全タグの統合
tags_all = []
tags_all.extend(weightscale.activity)
tags_all.extend(weightscale.positive)
tags_all.extend(weightscale.negative)
tags_all.extend(behavior.activity)
tags_all.extend(behavior.positive)
tags_all.extend(behavior.negative)
tags_all.extend(usertime.activity)
tags_all.extend(usertime.season)
tags_all.extend(usertime.positive)
tags_all.extend(usertime.negative)
tags_all.extend(usertime.others)
tags_all.extend(weather.info)
tags_all.extend(weather.feel)


print(tags_all)

#検索ワードの決定
search_word = []

#検索ワード数
num_words = 2

#検索ワードが1の場合は全ての単語から1つを選ぶ
if num_words == 1:
  search_word = random.sample(tags_all,1) 

#検索ワードが2の場合は全ての単語から2つを選ぶ
if num_words == 2:
  search_word = random.sample(tags_all,2)   

#検索ワードが2の場合は全ての単語から2つを選ぶ
if num_words == 3:
  search_word = random.sample(tags_all,3)   



#検索ワードが3の場合は行動/ポジティブ/ネガティブから1語づつランダムに選択する
#if num_words == 3:
#  search_word.append(weightscale.activity[0])
#  search_word.append(random.choice(weightscale.positive))
#  search_word.append(random.choice(weightscale.negative))



print(search_word)
```
